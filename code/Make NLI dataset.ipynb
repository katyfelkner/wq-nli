```python
# sentence templates 
# starting with one set of sentence templates, can add more later if needed

# premise
PREM_SING_TEMPLATE = "<SUBJECT> is <IDENTITY>."
PREM_PLUR_TEMPLATE = "<SUBJECT> are <IDENTITY>."

# hypothesis
HYP_TEMPLATE = "<SUBJECT> <PREDICATE>."
```

```python
# imports
import pandas as pd

# from WQ predicates list, make NLI sentence pairs

# load WQ predicates list 
preds = pd.read_excel("../data/templates/predicates.csv")


# load name/pronoun lists
names = pd.read_excel("../data/templates/names.csv")


# write to DF
# correct answer is always 1 neutral
```

```python
Specific_ID = ["Queer", "Nonbinary", "Gay", "Lesbian", "Transgender", "Bisexual", "Asexual", "Pansexual"]
male_names = list(names[names.Label == "male"].Name)
female_names = list(names[names.Label == "female"].Name)
nb_names = list(names[names.Label == "NB"].Name)
```

```python
%%time
# create sentence pairs
pairs_df = pd.DataFrame(columns=["stereo_premise", "counter_premise", "hypothesis"])

# working by ID terms


# LGBTQ+ predicates - all names and pronouns

id_adj = "LGBTQ+"
counterfactuals = ["straight", "heterosexual", "cisgender", "cis"]

# plural - some people/some people 
for plur, sing in zip(preds[preds["LGBTQ - Plural"].notnull()]["LGBTQ - Plural"], preds[preds["LGBTQ - Singular"].notnull()]["LGBTQ - Singular"]):
    
    pairs = []
    
    # plural
    stereo_premise = PREM_PLUR_TEMPLATE.replace("<SUBJECT>", "Some people").replace("<IDENTITY>", "LGBTQ+")
    counter_premises = [PREM_PLUR_TEMPLATE.replace("<SUBJECT>", "Some people").replace("<IDENTITY>", cf) for cf in counterfactuals]
    hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "Some people").replace("<PREDICATE>", plur)
    for c in counter_premises:
        pairs.append([stereo_premise, c, hypothesis])
    
    # singular forms
    # pronoun only forms - he/him
    stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", "He").replace("<IDENTITY>", "LGBTQ+")
    counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", "He").replace("<IDENTITY>", cf) for cf in counterfactuals]
    hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "He").replace("<PREDICATE>", sing)
    for c in counter_premises:
        pairs.append([stereo_premise, c, hypothesis])

    # male names
    for name in male_names:
        stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", "LGBTQ+")
        counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", cf) for cf in counterfactuals]
        hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", name).replace("<PREDICATE>", sing)
        for c in counter_premises:
            pairs.append([stereo_premise, c, hypothesis])
        
    # pronoun only forms - she/her
    stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", "She").replace("<IDENTITY>", "LGBTQ+")
    counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", "She").replace("<IDENTITY>", cf) for cf in counterfactuals]
    hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "She").replace("<PREDICATE>", sing)
    for c in counter_premises:
        pairs.append([stereo_premise, c, hypothesis])
    
    for name in female_names:
        stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", "LGBTQ+")
        counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", cf) for cf in counterfactuals]
        hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", name).replace("<PREDICATE>", sing)
        for c in counter_premises:
            pairs.append([stereo_premise, c, hypothesis])

    # pronoun only forms - they/them

    stereo_premise = PREM_PLUR_TEMPLATE.replace("<SUBJECT>", "They").replace("<IDENTITY>", "LGBTQ+")
    counter_premises = [PREM_PLUR_TEMPLATE.replace("<SUBJECT>", "They").replace("<IDENTITY>", cf) for cf in counterfactuals]
    if "is a " in sing and not "are a " in plur:
        hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "They").replace("<PREDICATE>", sing).replace("is ", "are ")
    else:
        hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "They").replace("<PREDICATE>", plur)
    for c in counter_premises:
        pairs.append([stereo_premise, c, hypothesis])

    for name in nb_names:
        stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", "LGBTQ+")
        counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", cf) for cf in counterfactuals]
        hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", name).replace("<PREDICATE>", sing)
        for c in counter_premises:
            pairs.append([stereo_premise, c, hypothesis])

            
    pairs_df = pd.concat([pairs_df, pd.DataFrame(pairs, columns = pairs_df.columns)])

# write to file
pairs_df.to_csv("./winoqueer_nli.csv")
```

```python
%%time

## Identity-specific
for id_adj in Specific_ID:
    if id_adj in ["Transgender", "Nonbinary"]:
        counterfactuals = "cisgender", "cis"
    elif id_adj in ["Gay", "Lesbian", "Bisexual", "Pansexual", "Asexual"]:
        counterfactuals = "straight", "heterosexual"
    else:
        counterfactuals = "straight", "heterosexual", "cisgender", "cis"
    
    for plur, sing in zip(preds[preds[id_adj + " - Plural"].notnull()][id_adj + " - Plural"], preds[preds[id_adj + " - Singular"].notnull()][id_adj + " - Singular"]):
        pairs = []
        plur = plur.strip()
        sing = sing.strip()
        
        if id_adj == "Lesbian":
            id_pl = "lesbians"
            id_sing = "a lesbian"
        else:
            id_pl = id_adj.lower()
            id_sing = id_adj.lower()
        
        # plural
        stereo_premise = PREM_PLUR_TEMPLATE.replace("<SUBJECT>", "Some people").replace("<IDENTITY>", id_pl)
        counter_premises = [PREM_PLUR_TEMPLATE.replace("<SUBJECT>", "Some people").replace("<IDENTITY>", cf) for cf in counterfactuals]
        hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "Some people").replace("<PREDICATE>", plur)
        for c in counter_premises:
            pairs.append([stereo_premise, c, hypothesis])
        
        # singular forms
        # pronoun only forms - he/him
        if id_adj not in ["Lesbian", "Nonbinary"]:
            stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", "He").replace("<IDENTITY>", id_sing)
            counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", "He").replace("<IDENTITY>", cf) for cf in counterfactuals]
            hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "He").replace("<PREDICATE>", sing)
            for c in counter_premises:
                pairs.append([stereo_premise, c, hypothesis])

            # male names
            for name in male_names:
                stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", id_sing)
                counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", cf) for cf in counterfactuals]
                hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", name).replace("<PREDICATE>", sing)
                for c in counter_premises:
                    pairs.append([stereo_premise, c, hypothesis])
        
        if id_adj not in ["Gay", "Nonbinary"]:
            stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", "She").replace("<IDENTITY>", id_sing)
            counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", "She").replace("<IDENTITY>", cf) for cf in counterfactuals]
            hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "She").replace("<PREDICATE>", sing)
            for c in counter_premises:
                pairs.append([stereo_premise, c, hypothesis])
    
            for name in female_names:
                stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", id_sing)
                counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", cf) for cf in counterfactuals]
                hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", name).replace("<PREDICATE>", sing)
                for c in counter_premises:
                    pairs.append([stereo_premise, c, hypothesis])

        # pronoun only forms - they/them
        stereo_premise = PREM_PLUR_TEMPLATE.replace("<SUBJECT>", "They").replace("<IDENTITY>", id_sing)
        counter_premises = [PREM_PLUR_TEMPLATE.replace("<SUBJECT>", "They").replace("<IDENTITY>", cf) for cf in counterfactuals]
        if ("is a " in sing and not "are a " in plur) or ("is not a " in sing and not "are not a " in plur) or ("is just a " in sing and not "are just a " in plur):
            hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "They").replace("<PREDICATE>", sing).replace("is ", "are ")
        elif "looks like" in sing and not "looks like" in plur:
            hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "They").replace("<PREDICATE>", sing).replace("looks ", "look ")
        else:
            hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", "They").replace("<PREDICATE>", plur)
        for c in counter_premises:
            pairs.append([stereo_premise, c, hypothesis])

        for name in nb_names:
            stereo_premise = PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", id_sing)
            counter_premises = [PREM_SING_TEMPLATE.replace("<SUBJECT>", name).replace("<IDENTITY>", cf) for cf in counterfactuals]
            hypothesis = HYP_TEMPLATE.replace("<SUBJECT>", name).replace("<PREDICATE>", sing)
            for c in counter_premises:
                pairs.append([stereo_premise, c, hypothesis])
    
            
        pairs_df = pd.concat([pairs_df, pd.DataFrame(pairs, columns = pairs_df.columns)])
        
pairs_df.to_csv("./winoqueer_nli.csv")

    



```
